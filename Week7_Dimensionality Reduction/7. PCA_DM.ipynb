### 유방암 예제 데이터 PCA 수행

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
#유방암 데이터 가져오기

from sklearn.datasets import load_breast_cancer
cancer=load_breast_cancer()
X = cancer.data
```

```python
pd.DataFrame(X)
```

```python
# Data Normalizing (Mean Centering)

from sklearn.preprocessing import StandardScaler
X_ = StandardScaler().fit_transform(X)
```

```python
# PCA 수행 - 주성분 2개

from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pc = pca.fit_transform(X_)
```

```python
#Plotting

import matplotlib.pyplot as plt
plt.scatter(pc[:,0],pc[:,1])
```

### 유방암 예제 데이터 PCA 수행 - 반응변수 활용하여 결과 확인 

```python
#유방암 데이터 가져오기

cancer=load_breast_cancer()
X = cancer.data
y = cancer.target
```

```python
# Data Normalizing (Mean Centering)

X_ = StandardScaler().fit_transform(X)
```

```python
# PCA 수행 - 주성분 2개

pca = PCA(n_components=2)
pc = pca.fit_transform(X_)
```

```python
# 카테고리정보 (음성 양성)

pc_y = np.c_[pc,y]
df = pd.DataFrame(pc_y,columns=['PC1','PC2','diagnosis'])
```

```python
#Plotting

import matplotlib.pyplot as plt
plt.scatter(x=df['PC1'],y=df['PC2'],c=df['diagnosis'])
```

```python
#Plotting

import seaborn as sns
sns.scatterplot(data=df,x='PC1',y='PC2',hue='diagnosis')
```

```python
# PCA 수행 - 주성분 3개
pca = PCA(n_components=3)
pc = pca.fit_transform(X_)

# 카테고리정보 (음성 양성)
pc_y = np.c_[pc,y]
df = pd.DataFrame(pc_y,columns=['PC1','PC2','PC3','diagnosis'])

#Plotting

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['PC1'], df['PC2'], df['PC3'], c=df['diagnosis'],  s=20)
```

```python
# PCA 수행 - 주성분 4개
pca = PCA(n_components=4)
pc = pca.fit_transform(X_)

df_var = pd.DataFrame({'var':pca.explained_variance_ratio_,
             'PC':['PC1','PC2','PC3','PC4']})
sns.barplot(x='PC',y="var", 
           data=df_var, color="c");
```
