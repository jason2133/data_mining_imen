1. Generating association rules from frequent itemsets

```python
#예제 데이터 생성

dataset=([['빵','우유'],['빵','기저귀','맥주','계란']
               ,['우유','기저귀','맥주','콜라'],['빵','우유','기저귀','맥주']
               , ['빵','우유','기저귀','콜라']])
```

```python
import pandas as pd
pd.DataFrame(data,columns={'itemset'})
```

```python
#!pip install mlxtend

from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, fpgrowth
```

```python
#데이터 셋 변환 

te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
#FP-Growth 알고리즘

fpg_frequent_itemsets = fpgrowth(df, min_support=0.5, use_colnames=True)
fpg_frequent_itemsets
```

```python
#Apriori 알고리즘

apr_frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)
apr_frequent_itemsets
```
