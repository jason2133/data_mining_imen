### MBTI 분류

```python
mbti_S = df[(df['type_label']==4) | (df['type_label']==5) | (df['type_label']==6) | (df['type_label']== 7) | (df['type_label']==12) | (df['type_label']==13) | (df['type_label']== 14) | (df['type_label']==15)]
mbti_N = df[(df['type_label']==0) | (df['type_label']==1) | (df['type_label']==2) | (df['type_label']== 3) | (df['type_label']==8) | (df['type_label']==9) | (df['type_label']== 10) | (df['type_label']==11)]

```

```python
from nltk.corpus import stopwords
stop = ['tapatalk', 'iphone', 'ha', 'using', 'enfjs', 'thats', 'kitteh', 'th', 'ya', 'alot', 'fucking', 'favourite', 'dont', 'sp', 'sx', 'dom', 'cant', 'esfjs', 'estps', 'enfps', 'entps', 'nf', 'sent', 'inferior', 'ive', 'score','particularly', 'youtube', 'husband', 'sensor', 'coffee', 'yea', 'idk', 'summer', 'fourm', 'welcome', 'much', 'istps', 'hahaha', 'wanna', 'istps', 'entjs', 'istjs', 'yep', 'typically', 'somebody', 'tmlt', 'mbti', 'type', 'intuition', 'typed', 'socionics','omg','as', 'intps', 'infps', 'sj', 'isfjs']

mbti_S['posts'] = mbti_S['posts'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
mbti_N['posts'] = mbti_N['posts'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
```

### DTM 생성

```python
#MBTI_S
from sklearn.feature_extraction.text import CountVectorizer
vec_S = CountVectorizer(max_df=0.1, max_features=1000, min_df=2, ngram_range=(1,2))
    
document_term_matrix_S = vec_S.fit_transform(mbti_S['posts'])
DTM_S = pd.DataFrame(document_term_matrix_S.toarray(), columns=vec_S.get_feature_names())  

#MBTI_N
vec_N = CountVectorizer(max_df=0.1, max_features=1000, min_df=2, ngram_range=(1,2))
document_term_matrix_N = vec_N.fit_transform(mbti_N['posts'])
DTM_N = pd.DataFrame(document_term_matrix_N.toarray(), columns=vec_N.get_feature_names())  
```

### TF-IDF 값 구하기

```python
from sklearn.feature_extraction.text import TfidfTransformer
transformer = TfidfTransformer()
transformer

MBTI_S_TFIDF = transformer.fit_transform(DTM_S)
print('#shape of tfidf matrix:', MBTI_S_TFIDF)

MBTI_N_TFIDF = transformer.fit_transform(DTM_N)
print('#shape of tfidf matrix:', MBTI_N_TFIDF)
```

```python
MBTI_S_TFIDF_DF = pd.DataFrame(MBTI_S_TFIDF.toarray(), columns=vec_S.get_feature_names())
MBTI_S_TFIDF_DF

MBTI_N_TFIDF_DF = pd.DataFrame(MBTI_N_TFIDF.toarray(), columns=vec_N.get_feature_names())
MBTI_N_TFIDF_DF
```

```python
MBTI_S_TFIDF_SUM = MBTI_S_TFIDF_DF.sum()
MBTI_S_TFIDF_SUM.sort_values(ascending=False)[0:50]

```

```python
MBTI_N_TFIDF_SUM = MBTI_N_TFIDF_DF.sum() 
MBTI_N_TFIDF_SUM.sort_values(ascending=False)[0:50]
```

### LDA분석


```python
from sklearn.decomposition import LatentDirichletAllocation



document_term_matrix_S
print('CountVectorizer Shape:', document_term_matrix_S.shape)

lda = LatentDirichletAllocation(n_components=2)
lda.fit(document_term_matrix_S)

def display_topics(model, feature_names, num_top_words):
    for topic_index, topic in enumerate(model.components_):
        print('Topic #', topic_index)

        topic_word_indexes = topic.argsort()[::-1]
        top_indexes=topic_word_indexes[:num_top_words]
        
        feature_concat = ' '.join([feature_names[i] for i in top_indexes])                
        print(feature_concat)

feature_names = vec_S.get_feature_names()

display_topics(lda, feature_names, 10)
```

```python
document_term_matrix_N
print('CountVectorizer Shape:', document_term_matrix_N.shape)

lda_N = LatentDirichletAllocation(n_components=2)
lda_N.fit(document_term_matrix_N)

feature_names = vec_N.get_feature_names()

display_topics(lda_N, feature_names, 10)
```
