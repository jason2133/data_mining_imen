### MBTI 분류하기

```python
mbti_T = df[(df['type_label']==2) | (df['type_label']==3) | (df['type_label']==6) | (df['type_label']== 7) | (df['type_label']==10) | (df['type_label']==11) | (df['type_label']== 14) | (df['type_label']==15)]
mbti_F = df[(df['type_label']==0) | (df['type_label']==1) | (df['type_label']==4) | (df['type_label']== 5) | (df['type_label']==8) | (df['type_label']==9) | (df['type_label']== 12) | (df['type_label']==13)]
```

### 추가적인 불용어 처리

```python
from nltk.corpus import stopwords
stop = ['tapatalk', 'iphone', 'ha', 'using', 'enfjs', 'thats', 'kitteh', 'th', 'ya', 'alot', 'fucking', 'favourite', 'dont', 'sp', 'sx', 'dom', 'cant', 'esfjs', 'estps', 'enfps', 'entps', 'nf', 'sent', 'inferior', 'ive', 'score','particularly', 'youtube', 'husband', 'sensor', 'coffee', 'yea', 'idk', 'summer', 'fourm', 'welcome', 'much', 'istps', 'hahaha', 'wanna', 'istps', 'entjs', 'istjs', 'yep', 'typically', 'somebody', 'tmlt', 'mbti', 'type', 'intuition', 'typed', 'socionics','omg','as', 'intps', 'infps', 'sj', 'isfjs', 'xd', 'nt', 'isfps']

mbti_T['posts'] = mbti_T['posts'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
mbti_F['posts'] = mbti_F['posts'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
```

### DTM 생성

```python
#MBTI_T
from sklearn.feature_extraction.text import CountVectorizer
vec_T = CountVectorizer(max_df=0.1, max_features=1000, min_df=2, ngram_range=(1,2))
    
document_term_matrix_T = vec_T.fit_transform(mbti_T['posts']) 
DTM_T = pd.DataFrame(document_term_matrix_T.toarray(), columns=vec_T.get_feature_names())  

#MBTI_F
vec_F = CountVectorizer(max_df=0.1, max_features=1000, min_df=2, ngram_range=(1,2))
document_term_matrix_F = vec_F.fit_transform(mbti_F['posts'])
DTM_F = pd.DataFrame(document_term_matrix_F.toarray(), columns=vec_F.get_feature_names())  
```

### TF-IDF 값 구하기

```python
from sklearn.feature_extraction.text import TfidfTransformer
transformer = TfidfTransformer()
transformer

MBTI_T_TFIDF = transformer.fit_transform(DTM_T)
print('#shape of tfidf matrix:', MBTI_T_TFIDF)

MBTI_F_TFIDF = transformer.fit_transform(DTM_F)
print('#shape of tfidf matrix:', MBTI_F_TFIDF)
```

```python
MBTI_T_TFIDF_DF = pd.DataFrame(MBTI_T_TFIDF.toarray(), columns=vec_T.get_feature_names())
MBTI_T_TFIDF_DF

MBTI_F_TFIDF_DF = pd.DataFrame(MBTI_F_TFIDF.toarray(), columns=vec_F.get_feature_names())
MBTI_F_TFIDF_DF
```

```python
MBTI_T_TFIDF_SUM = MBTI_T_TFIDF_DF.sum()
MBTI_T_TFIDF_SUM.sort_values(ascending=False)[0:50]
```

```python
MBTI_F_TFIDF_SUM = MBTI_F_TFIDF_DF.sum() 
MBTI_F_TFIDF_SUM.sort_values(ascending=False)[0:50]
```

### LDA 분석

```python
from sklearn.decomposition import LatentDirichletAllocation

document_term_matrix_T
print('CountVectorizer Shape:', document_term_matrix_T.shape)

lda = LatentDirichletAllocation(n_components=2)
lda.fit(document_term_matrix_T)

def display_topics(model, feature_names, num_top_words):
    for topic_index, topic in enumerate(model.components_):
        print('Topic #', topic_index)


        topic_word_indexes = topic.argsort()[::-1]
        top_indexes=topic_word_indexes[:num_top_words]
        

        feature_concat = ' '.join([feature_names[i] for i in top_indexes])                
        print(feature_concat)


feature_names = vec_T.get_feature_names()


display_topics(lda, feature_names, 10)
```

```python
document_term_matrix_F
print('CountVectorizer Shape:', document_term_matrix_F.shape)

lda_F = LatentDirichletAllocation(n_components=2)
lda_F.fit(document_term_matrix_F)


feature_names = vec_F.get_feature_names()


display_topics(lda_F, feature_names, 10)
```
